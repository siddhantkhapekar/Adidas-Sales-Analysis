```python
df["price_per_unit"]=df["price_per_unit"].fillna(df["price_per_unit"].mean())
```

```python
df["units_sold"]=df["units_sold"].fillna(df["units_sold"].mean())
```

```python
df["total_sales"]=df["total_sales"].fillna(df["total_sales"].mean())
```

```python
df['monthly_sales'] = pd.to_numeric(df['monthly_sales'], errors='coerce')
df["monthly_sales"]=df["monthly_sales"].fillna(df["monthly_sales"].mean())
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python

total_sales_per_customer = df.groupby('customer')['total_sales'].sum().reset_index()

plt.figure(figsize=(10, 6))
c=["r","k","y","b","g","c"]
plt.bar(total_sales_per_customer['customer'], total_sales_per_customer['total_sales'],color=c)
plt.xlabel('Customer')
plt.ylabel('Total Sales Amount')
plt.title('Total Sales per Customer')
plt.show()
```

```python

df['invoice_date'] = pd.to_datetime(df['invoice_date'], format='%d-%m-%Y')
df.set_index('invoice_date', inplace=True)
df.sort_index(inplace=True)
monthly_sales = df['total_sales'].resample('M').sum()
monthly_sales_ma = monthly_sales.rolling(window=3).mean()


plt.figure(figsize=(12, 6))
plt.plot(monthly_sales, label='Monthly Sales', marker='o')
plt.plot(monthly_sales_ma, label='3-Month Moving Average', linestyle='--')
plt.title('Monthly Sales and 3-Month Moving Average')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.legend()
plt.grid(True)
plt.show()
```
